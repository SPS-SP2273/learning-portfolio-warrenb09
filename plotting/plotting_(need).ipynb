# A simple plot

## Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='--')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

From the simple plot above, we can see different basic syntaxes of `pyplot` functions.

|**Fucntion**|**Meaning**|**Basic syntax**|
|:--:|:--:|:--:|
|`plt.plot`|Plots the graph and gives the graph some attributes|`plt.plot(x-values, y-values)`|
|`plt.xlabel`|Names the x-axis|`plt.xlabel('label')`|
|`plt.ylabel`|Names the y-axis|`plt.ylabel('label')`|
|`plt.title`|Gives the whole plot a title|`plt.title('title')`|
|`plt.grid`|Gives the plot grid lines|`plt.grid(True)`|
|`plt.legend`|Specifies the location of labels of `plt.plot`|`plt.legend(loc = 'position')`|
|`plt.show`|Shows the plot (important in non-interactive environments)|`plt.show()`|

<p></p>

Notice also there are additional parameters in addition to the basic syntax. Below shows the parameters and examples.

|**Parameter**|**Meaning**|**Example**|
|:--:|:--:|:--:|
|`colour` or `c`|Sets the colour of lines or words|`c = 'red'`|
|`alpha`|Sets the transparency of lines|`alpha = 0.5`|
|`linewidth` or `lw`|Sets the thickness of lines|`lw = 2`|
|`linestyle` or `ls`|Sets the style of lines|`ls = '-'`; `-` for solid line, `--` for dashed line, `:` for dotted line, `-.` for dash-dot line.|
|`marker`|Sets the style of points for plotted lines|`marker = 'o'`; `o` for circle, `s` for square, `D` for diamond, `^` for triangle, etc.|

There are more parameters that can be found [here](https://matplotlib.org/stable/api/markers_api.html).

*Note that the order of parameters in the function does not matter. So, `plt.plot(x, y, c = 'red', alpha = 0.5)` is the same as `plt.plot(x, y, alpha = 0.5, c = 'red')`.

## Adding another plot

We can add another plot to an existing plot. Simply use another `plt.plot` function.

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='--')
plt.plot(x, y_2,
         label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

## Yet another plot but with error bars

We can plot a line with error bars using the `plt.errorbar` function. The basic syntax is `plt.errorbar(x-values, y-values, xerr = xerror, yerr = yerror)`. The `xerror` and `yerror` must be provided as numbers.

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='--')
plt.plot(x, y_2,
         label='Y2 values', color='blue')
plt.errorbar(x, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

Here, the errors for all values of x and y are constant, but we can also give a list of errors for different errors. See below.

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_error = [.3, .2, .15, .6, .1, .4, .75]

plt.errorbar(x, y_1, label='Y values',
         color='red', linestyle='--', yerr = y_error)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

# Better with NumPy

For plotting, it is easier to use `numpy` arrays rather than standard Python arrays. The benefits of `numpy` arrays are shown below.

## Adding mathematical functions

With `numpy` arrays it is easy to plot mathematical functions over a range of values of x. For example, we want to plot the graph $y = x^{2}$ and $y = sin(x)$ over the range $0 \leq x \leq 6$.

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5, 6])       #set range of x
y1 = x ** 2
y2 = np.sin(x)

plt.plot(x, y1, label = "y = x^2", ls = "--")
plt.plot(x, y2, label = "y = sin(x)")
plt.grid(alpha = .25)
plt.legend()
```

At the moment, the graph of $y=sin(x)$ is very compressed because the range of values of $y$ for it is $\pm 1$, while $y=x^{2}$ does not have that range restriction. We can solve this by making another y-axis that individually correspond to each of the y-values of the two different graphs.

## We need another axis!

We can create another y-axis that shares the same x-axis using the `plt.twinx()` function.

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5, 6])       #set range of x
y1 = x ** 2
y2 = np.sin(x)

plt.plot(x, y1, label = "y = x^2", ls = "--")
plt.legend()                              #remember this to create separate legends for each plot

plt.twinx()                               #creates a new y-axis for the plot that comes after

plt.plot(x, y2, label = "y = sin(x)", color = "r")
plt.grid(alpha = .25)
plt.legend()
```

## Another case of using NumPy

Our plot still looks ugly because it is not a smooth curve even though they are curvilinear lines. We can fix this using the `np.linspace` function.

The `np.linspace` function creates a specified number of evenly spaced numbers in a list. The syntax is `np.linspace(start, stop, number of evenly spaced numbers to generate)`. See below for an example.

```python
np.linspace(0, 1, 5)                    #start number is 0, end number is 1, generating 5 evely spaced numbers
```

So, for plotting, `np.linspace` can create large amounts of x-values to be plotted to give a smooth curve.

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 6, 500)                #set range of x, with 500 x-values
y1 = x ** 2
y2 = np.sin(x)

plt.plot(x, y1, label = "y = x^2", ls = "--")
plt.legend()                              

plt.twinx()                             

plt.plot(x, y2, label = "y = sin(x)", color = "r")
plt.grid(alpha = .25)
plt.legend()
```

# Saving to disc

If we want to use the plot for a report or presentation, we must save it first. We can easily do this using the function `plt.savefig`.

The syntax is `plt.savefig('file name with extension', dpi = resolution)`.

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 6, 500)                #set range of x, with 500 x-values
y1 = x ** 2
y2 = np.sin(x)

plt.plot(x, y1, label = "y = x^2", ls = "--")
plt.legend()                              

plt.twinx()                             

plt.plot(x, y2, label = "y = sin(x)", color = "r")
plt.grid(alpha = .25)
plt.legend()

plt.savefig('testplot.png', dpi = 600)                #saves in the current directory
```

# A real example: Global Warming

## Plotting data from files

To plot data from files, we use the `numpy` package to load the data into Python first. We use the `np.loadtxt` function.

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)                      #skiprows to prevent the first 24 rows (no data) from being loaded

#data is in the form of a 2D list, with the 1D list elements being a row of data
date = data[:, 0]                                    #the : in [:, 0] means we are considering all the 1D lists for plotting, and the 0 extracts the first element, which is the date
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-3, 3])                                    #make the y-axis have a range of -3 to 3

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')     #creates a horizontal line
```

There are two new functions included in the above code:
- `plt.ylim`, whose syntax is `plt.ylim([min y, max y])`
*note that `plt.xlim` also exists with a similar syntax
- `plt.hlines`, whose syntax is `plt.hlines(y value, min x, max x)`
*note that `plt.vlines` also exists with a similar syntax

## Styles

There are some [styles](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html) that we might want to use for our plots (but honestly the standard styles are ugly).

We can use the styles by using the function `plt.style.use(name of style)`. Below shows an example.

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)                      

date = data[:, 0]                                    
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-3, 3])
plt.title("Temperature anomalies from 1880 to 2023\n(relative to averages from Jan 1951 to Dec 1980)")
plt.xlabel("Date")
plt.ylabel("Temperature anomaly")
plt.grid(alpha = .25)

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') 
```
