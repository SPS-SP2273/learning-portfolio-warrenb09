## 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='--')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

From the simple plot above, we can see different basic syntaxes of `pyplot` functions.

|**Fucntion**|**Meaning**|**Basic syntax**|
|:--:|:--:|:--:|
|`plt.plot`|Plots the graph and gives the graph some attributes|`plt.plot(x-values, y-values)`|
|`plt.xlabel`|Names the x-axis|`plt.xlabel('label')`|
|`plt.ylabel`|Names the y-axis|`plt.ylabel('label')`|
|`plt.title`|Gives the whole plot a title|`plt.title('title')`|
|`plt.grid`|Gives the plot grid lines|`plt.grid(True)`|
|`plt.legend`|Specifies the location of labels of `plt.plot`|`plt.legend(loc = 'position')`|
|`plt.show`|Shows the plot (important in non-interactive environments)|`plt.show()`|

<p></p>

Notice also there are additional parameters in addition to the basic syntax. Below shows the parameters and examples.

|**Parameter**|**Meaning**|**Example**|
|:--:|:--:|:--:|
|`colour` or `c`|Sets the colour of lines or words|`c = 'red'`|
|`alpha`|Sets the transparency of lines|`alpha = 0.5`|
|`linewidth` or `lw`|Sets the thickness of lines|`lw = 2`|
|`linestyle` or `ls`|Sets the style of lines|`ls = '-'`; `-` for solid line, `--` for dashed line, `:` for dotted line, `-.` for dash-dot line.|
|`marker`|Sets the style of points for plotted lines|`marker = 'o'`; `o` for circle, `s` for square, `D` for diamond, `^` for triangle, etc.|

There are more parameters that can be found [here](https://matplotlib.org/stable/api/markers_api.html).

*Note that the order of parameters in the function does not matter. So, `plt.plot(x, y, c = 'red', alpha = 0.5)` is the same as `plt.plot(x, y, alpha = 0.5, c = 'red')`.

## 1.2 Adding another plot

We can add another plot to an existing plot. Simply use another `plt.plot` function.

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='--')
plt.plot(x, y_2,
         label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

## 1.3 Yet another plot but with error bars

We can plot a line with error bars using the `plt.errorbar` function. The basic syntax is `plt.errorbar(x-values, y-values, xerr = xerror, yerr = yerror)`. The `xerror` and `yerror` must be provided as numbers.

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, label='Y values',
         color='red', linestyle='--')
plt.plot(x, y_2,
         label='Y2 values', color='blue')
plt.errorbar(x, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```
